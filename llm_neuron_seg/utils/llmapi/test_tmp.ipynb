```python
import os
import logging
import requests

# 确保代理设置正确
os.environ['HTTP_PROXY'] = 'http://127.0.0.1:7890'
os.environ['HTTPS_PROXY'] = 'https://127.0.0.1:7890'

# 启用日志记录
logging.basicConfig(level=logging.DEBUG)
logger = logging.getLogger()

try:
    logger.info("Attempting to connect to OpenAI API...")
    response = requests.get('https://api.openai.com/v1/chat/completions',
                            headers={
                                "Authorization": f"Bearer {os.environ['OPENAI_API_KEY']}"
                            })
    response.raise_for_status()
    print(response.json())
except requests.exceptions.ProxyError as proxy_error:
    logger.error("Proxy error occurred: %s", proxy_error)
except requests.exceptions.SSLError as ssl_error:
    logger.error("SSL error occurred: %s", ssl_error)
except requests.exceptions.RequestException as e:
    logger.error("Request error occurred: %s", e)

# 不使用代理进行测试
try:
    logger.info("Attempting to connect to OpenAI API without proxy...")
    os.environ.pop('HTTP_PROXY', None)
    os.environ.pop('HTTPS_PROXY', None)
    response = requests.get('https://api.openai.com/v1/chat/completions',
                            headers={
                                "Authorization": f"Bearer {os.environ['OPENAI_API_KEY']}"
                            })
    response.raise_for_status()
    print(response.json())
except requests.exceptions.RequestException as e:
    logger.error("An error occurred when not using proxy: %s", e)
```
